```python

# Function to categorize calories
def categorize_calories(calories):
    if calories == 0:
        return "no-calories"
    elif calories < 200:
        return "low-calorie"
    elif calories < 400:
        return "mid-calorie"
    else:
        return "high-calorie"


# Function to categorize serving size (in grams or milliliters)
def categorize_serving_size(serving_size):
    print("=======================")
    print(serving_size)
    print("=======================")
    # size = int(serving_size.strip())
    return "lakalaka"
    # if size == 0:
    #     return "no-serving-size"
    # elif size < 100:
    #     return "small-portion"
    # elif size < 300:
    #     return "medium-portion"
    # else:
    #     return "large-portion"


# Function to categorize fat levels
def categorize_fat(fat_g):
    if fat_g == 0:
        return "fat-free"
    elif fat_g < 3:
        return "low-fat"
    elif fat_g < 10:
        return "mid-fat"
    else:
        return "high-fat"


# Function to categorize saturated fat levels
def categorize_sat_fat(sat_fat_g):
    if sat_fat_g == 0:
        return "sat-fat-free"
    elif sat_fat_g < 2:
        return "low-sat-fat"
    elif sat_fat_g < 5:
        return "mid-sat-fat"
    else:
        return "high-sat-fat"


# Function to categorize cholesterol levels
def categorize_cholesterol(cholesterol_mg):
    if cholesterol_mg == 0:
        return "cholesterol-free"
    elif cholesterol_mg < 100:
        return "low-cholesterol"
    elif cholesterol_mg < 200:
        return "mid-cholesterol"
    else:
        return "high-cholesterol"


# Function to categorize sodium levels
def categorize_sodium(sodium_mg):
    if sodium_mg == 0:
        return "sodium-free"
    elif sodium_mg < 150:
        return "low-sodium"
    elif sodium_mg < 400:
        return "mid-sodium"
    else:
        return "high-sodium"


# Function to categorize carbohydrates levels
def categorize_carbs(carbohydrates_g):
    if carbohydrates_g == 0:
        return "carb-free"
    elif carbohydrates_g < 15:
        return "low-carb"
    elif carbohydrates_g < 50:
        return "mid-carb"
    else:
        return "high-carb"


# Function to categorize sugar levels
def categorize_sugar(sugar_g):
    if sugar_g == 0:
        return "sugar-free"
    elif sugar_g < 5:
        return "low-sugar"
    elif sugar_g < 15:
        return "mid-sugar"
    else:
        return "high-sugar"


# Function to categorize fiber levels
def categorize_fiber(fiber_g):
    if fiber_g == 0:
        return "fiber-free"
    elif fiber_g < 3:
        return "low-fiber"
    elif fiber_g < 8:
        return "mid-fiber"
    else:
        return "high-fiber"


# Function to categorize protein levels
def categorize_protein(protein_g):
    if protein_g == 0:
        return "protein-free"
    elif protein_g < 5:
        return "low-protein"
    elif protein_g < 20:
        return "mid-protein"
    else:
        return "high-protein"


available_entities = [
    "no-calories",
    "low-calorie",
    "mid-calorie",
    "high-calorie",
    "no-serving-size",
    "small-portion",
    "medium-portion",
    "large-portion",
    "fat-free",
    "low-fat",
    "mid-fat",
    "high-fat",
    "sat-fat-free",
    "low-sat-fat",
    "mid-sat-fat",
    "high-sat-fat",
    "cholesterol-free",
    "low-cholesterol",
    "mid-cholesterol",
    "high-cholesterol",
    "sodium-free",
    "low-sodium",
    "mid-sodium",
    "high-sodium",
    "carb-free",
    "low-carb",
    "mid-carb",
    "high-carb",
    "sugar-free",
    "low-sugar",
    "mid-sugar",
    "high-sugar",
    "fiber-free",
    "low-fiber",
    "mid-fiber",
    "high-fiber",
    "protein-free",
    "low-protein",
    "mid-protein",
    "high-protein",
]
```

```python

df["entities"] = df.apply(
    lambda x: [
        categorize_calories(x["calories"]),
        categorize_fat(x["fat_g"]),
        categorize_sat_fat(x["sat_fat_g"]),
        categorize_cholesterol(x["cholesterol_mg"]),
        categorize_sodium(x["sodium_mg"]),
        categorize_carbs(x["carbohydrates_g"]),
        categorize_sugar(x["sugar_g"]),
        categorize_fiber(x["fiber_g"]),
        categorize_protein(x["protein_g"])
    ],
    axis=1,
)
```

```python
from ast import literal_eval
def categorize_serving_size(entities, serving_size):
    # Determine the portion size value
    if serving_size == 0:
        val = "no-serving-size"
    elif serving_size < 100:
        val = "small-portion"
    elif serving_size < 300:
        val = "medium-portion"
    else:
        val = "large-portion"
    
    # Update the entities list
    entities.append(val)
    return entities

# Apply the function and update the DataFrame
df['entities'] = df.apply(lambda x: categorize_serving_size(literal_eval(x['entities']), x['serving_size']), axis=1)
```
